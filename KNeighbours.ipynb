```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.model_selection import train_test_split as TTS
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
data = pd.read_csv("../../Downloads/7th data.csv", names=list(pd.read_csv("../../Downloads/7th attribute names.csv").columns))
data.head(n=1)
```

```python
X = data.drop('heartdisease', axis = 'columns', inplace=False)
y = data['heartdisease']
print(set(y))
Xtrain, Xtest, ytrain, ytest = TTS(X, y, test_size=0.25, random_state=70)
```

```python
for i in range(3, 11):  
    clf = KNN(n_neighbors=i, algorithm='auto', metric='minkowski')
    clf.fit(Xtrain, ytrain)
    ypred = clf.predict(Xtest)
    print('No. of neighbors = ',i)
    print('Classifier accuracy details : \n', classification_report(ytest, ypred))
    print('Accuracy = ',accuracy_score(ytest, ypred)*100, '%', end='\n\n')
    print('Confusion matrix : \n', confusion_matrix(ytest, ypred))
    print('\n\n')
```
